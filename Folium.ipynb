Getting started with Folium; a Python wrapper for Leaflet.js 

```python
# Loading the required libraries
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
# Defining the coordinates of the center of our map
nbo_coords = [-1.29, 36.82]

# Creating the map
nbo_map = folium.Map(location = nbo_coords, zoom_start = 13)

# Display the map
nbo_map
```

Adding markers to our map

```python
# Defining the coordinates of our markers
Britam_tower_coords = [-1.2971424017095319, 36.813182931123535]
Family_basilica_coords = [-1.2847644863281125, 36.8198589915004]
Archives_coords = [-1.2847525061129854, 36.82588334598308]

# Adding markers to the map
folium.Marker(Britam_tower_coords, popup= 'Britam Tower').add_to(nbo_map)
folium.Marker(Family_basilica_coords, popup = 'Holy Familica Basilica').add_to(nbo_map)
folium.Marker(Archives_coords, popup = 'Kenya National Archives').add_to(nbo_map)

# Displaying the map
nbo_map
```

```python
# Changing the tiles since OSM is the default
# from branca.element import Figure 
# # branca is a spinoff from folium that hosts non-map specific features
# # we are using it here to resize the map 

# fig = Figure(width=550,height=350)
# fig.add_child(nbo_map)

# Adding other tile layers
folium.TileLayer('Stamen Terrain').add_to(nbo_map)
folium.TileLayer('Stamen Toner').add_to(nbo_map)
folium.TileLayer('Stamen Water Color').add_to(nbo_map)
folium.TileLayer('cartodbpositron').add_to(nbo_map)
folium.TileLayer('cartodbdark_matter').add_to(nbo_map)
folium.LayerControl().add_to(nbo_map)
nbo_map
```

```python
# Adding custom base maps
# Creating a dictionary to hold the basemaps
basemap = {
    'Google Maps' : folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=m&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Maps',
        overlay = True,
        control = True
    ),
    'Google Satellite Hybrid' : folium.TileLayer(
        tiles = 'https://mt1.google.com/vt/lyrs=y&x={x}&y={y}&z={z}',
        attr = 'Google',
        name = 'Google Satellite',
        overlay = True,
        control = True
    )
}

# Adding the custom base layers
ke_map = folium.Map(location=[-0.176, 37.91])
basemap['Google Maps'].add_to(ke_map)
ke_map
```
